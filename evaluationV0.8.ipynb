# Introduction

Identifying individuals in a group of animals is an active research topic as it paves the way to behavioural studies for lameness and illness assessment. Real-world applications demand algorithms to identify previously unseen individuals. For instance, traditional classification algorithms such as Convolutional Neural Networks, which must be trained exclusively for every cow at a farm, cannot classify or differentiate newly added cattle without prior re-training. Metric learning is a solution where deep learning models learn a latent space of differentiating features (or embedding) as a distance metric for simple clustering algorithms to associate similar individuals.
In this project, we independently replicated the deep metric learning technique proposed by Andrew et al. (2020)[<sup>[1]</sup>](#fn1) to train a Residual Convolutional Neural Network (Resnet) on a dataset of Holstein-Friesian cattle, which can identify unknown individuals, previously unseen during training. The dataset we used for training the Resnet is called [OpenCows2020](https://data.bris.ac.uk/data/dataset/10m32xl88x2b61zlkkgz3fml17) [<sup>[2]</sup>](#fn2) which was released by the authors under [Non-Commercial Government Licence for public sector information](http://www.nationalarchives.gov.uk/doc/non-commercial-government-licence/non-commercial-government-licence.htm). The technique relies on learning a 128-dimensional latent feature space that, when used to fit simple algorithms such as K-Nearest Neighbours, can perform open-set identification by clustering identical cows. We also implemented the proposed hybrid loss function and compared its performance with traditional losses often used in identification tasks. Furthermore, we also assessed the quality of the 128-dimensional embedding trained on the proposed loss function vs the traditional losses by t-SNE dimensionality reduction.

> üìì **This is a faithful (from scratch) implementation of the original paper focusing on using the PyTorch built-in functions. The original repository associated with the Andrew et al. paper[<sup>[1]</sup>](#fn1) uses a custom (although identical) ResNet `nn.Module`, whereas here, we directly rely on what PyTorch can offer.**

## ResNet based model for deep metric learning
Many real-world problems require an identification algorithm to work with examples unseen or withheld during training. In their study, Andrew et al. (2020) [<sup>[1]</sup>](#fn1) trained the Resnet-50 model (See Figure 1) to learn a latent space that clusters images of identical individuals with good separation from others. In theory, unknown individual embeddings (ones not seen during the training) should be significantly distant if the model had learnt discriminative features, eliminating the need for re-training to account for new individuals. The fully connected layer "FC 128" outputs a 128-dimensional embedding, which can be used as a distance metric with clustering algorithms to represent input similarity, hence the metric learning term. 

In this study, we assessed the generalisation capability of the proposed model by varying the number of unseen examples during training and monitoring the accuracy of the K-nearest neighbours on the resulting embeddings. For instance, in the Figure from the previous cell, we withhold 25% examples; therefore, the distribution is 25% open.

### Construction
The study used the Resnet50 [<sup>[3]</sup>](#fn3) model with pre-trained weight from Imagenet, which can be found [here](https://pytorch.org/vision/stable/models.html) (we used `resnet50-19c8e357.pth`). The authors used the Resnet-50 variant, which has four stages of residual blocks (See Figure 1 below). Fine-tuning a deep network like Resnet-50 is computationally expensive and can lead to premature convergence without prior hyper-parameterisation. Contrary to the author's approach of fine-tuning the entire model, we chose *not* to update the residual layers (see "Frozen", Figure below) and trained just the last three fully-connected layers ("Trained", Figure below) to speed up the learning process. Furthermore, we employed Pytorch's built-in methods to implement the model rather than implementing the entire Resnet model from scratch, as done in Andrew et al. (2020)'s [repository](https://github.com/CWOA/MetricLearningIdentification).

<center><img width="1000" alt="portfolio_view" src="https://github.com/Asheeshkrsharma/Identification-OpenCows/blob/main/staticAssets/Model_horizontal.svg?raw=1"><p>Figure 1: Resnet-50 model</p></center>

````python
class ResNetMod(nn.Module):
    def __init__(self, numClasses, embeddingSize=128):
        """Initialize the model

        Parameters
        ----------
        numClasses : Int
            Number of classes (or individuals in a dataset)
        embeddingSize : Int, optional
            Size of the final embedding used differentiate individuals

        Returns
        -------
        A ResNetMod class instance
        """
        super(ResNetMod, self).__init__()

        # We create a resnet50 block and load imagenet weight into it
        self.backbone = models.resnet50(pretrained=True)
        imagenetWeights = model_zoo.load_url(
            "https://download.pytorch.org/models/resnet50-19c8e357.pth"
        )
        self.backbone.load_state_dict(imagenetWeights)

        # These layers are trained again, bias is set to true.
        self.backbone.fc = nn.Linear(2048, 1000)
        self.relu = nn.ReLU(inplace=True)
        self.embedding = nn.Linear(1000, embeddingSize)
        self.softmax = nn.Linear(1000, numClasses)

    def forwardOnce(self, x):
        """Forward through the network once

        Parameters
        ----------
        x : Input image to embed
            [Batch Size, Channel, Width, Height]

        Returns
        -------
        Embedding
            [Batch Size, 128]
        Softmax
            [Batch Size, Num Classes]
        """

        # Conv1 -> Relu(bn1) -> maxpool
        x = self.backbone.maxpool(
            self.backbone.relu(self.backbone.bn1(self.backbone.conv1(x)))
        )
        # avgpool(Layer 1 -> Layer 2 -> Layer 3 -> Layer 4)
        x = self.backbone.avgpool(
            self.backbone.layer4(
                self.backbone.layer3(self.backbone.layer2(self.backbone.layer1(x)))
            )
        )
        # Relu(FC)
        x = self.relu(self.backbone.fc(x.view(x.size(0), -1)))
        return self.embedding(x), self.softmax(x)

    def forward(self, anchor, negative=None, positive=None):
        """Forward the network

        Parameters
        ----------
        anchor : Input image to embed
            [Batch Size, Channel, Width, Height]
        negative : Input image, optional
            [Batch Size, Channel, Width, Height] used during
            training with Triplet loss
        positive : Input image, optional
            [Batch Size, Channel, Width, Height] used during
            training with Triplet loss

        Returns
        -------
        list
            Embeddings and softmax ouputs for negative, anchor,
            and postive triplets in case nagative and postive
            image inputs are provided. Otherwise, returns a
            single embedding and softmax output.
        """
        if (negative != None) & (positive != None):
            negativeEMBD, negativeSMax = self.forwardOnce(negative)
            anchorEMBD, anchorSMax = self.forwardOnce(anchor)
            positiveEMBD, positiveSMax = self.forwardOnce(positive)
            return (
                negativeEMBD,
                anchorEMBD,
                positiveEMBD,
                torch.cat((negativeSMax, anchorSMax, positiveSMax), 0),
            )
        else:
            anchorEMBD, anchorSMax = self.forwardOnce(anchor)
            return anchorEMBD, anchorSMax

    def __repr__(self):
        summary = f"{self.backbone.conv1.__repr__()} \n{self.backbone.bn1.__repr__()}"
        summary += "\nResnet Backbone: \n"
        layers = [
            self.backbone.layer1,
            self.backbone.layer2,
            self.backbone.layer3,
            self.backbone.layer4,
        ]
        for layer in layers:
            parameters = list(layer.named_parameters())
            summary += (
                f"\t ‚é£ Input: {list(parameters[0][1].size())}"
                + f" -> Output: {list(parameters[-1][1].size())} ‚Ü©\n"
            )
        summary += self.backbone.fc.__repr__()
        summary += "\nOutput 1: " + self.embedding.__repr__()
        summary += "\nOutput 2: " + self.softmax.__repr__()
        return summary
````

### Loss functions
Discussion on loss functions quoted from the Andrew Et al. (2020) [<sup>[1]</sup>](#fn1) metric learning paper.

> $\cdots$ The problem with this [contrastive loss] formulation is that it cannot simultaneously encourage learning of visual similarities and dissimilarities, both of which are critical for obtaining clean, well-separated clusters on our coat pattern differentiation task. This shortcoming can be overcome by a triplet loss formulation; utilising the embeddings $x_a$, $x_p$, $x_n$ of a triplet containing three image inputs ($X_a$,$X_p$,$X_n$) denoting an anchor, a positive example from the same class, and a negative example from a different class, respectively. The idea being to encourage minimal distance between the anchor $x_a$ and the positive $x_p$, and maximal distance between the anchor $x_a$ and the negative sample $x_n$ in the embedded space. The loss function is given by:
> $$ \mathbb{L}_{TL} = max(0, d(x_a, x_p) - d(x_a, x_n) + \alpha)$$
> where $\alpha$ denotes a constant margin hyperparameter. The inclusion of the constant $\alpha$ often turns out to cause training issues since the margin can be satisfied at any distance from the anchor $\cdots$ a recent formulation named reciprocal triplet loss alleviates this limitation, which removes the margin hyperparameter altogether:
> $$ \mathbb{L}_{RTL} = d(x_a, x_p) + \frac{1}{d(x_a, x_n)} $$
> Recent work has demonstrated improvements in open-set recognition on various datasets via the inclusion of a SoftMax term in the triplet loss formulation during training given by:
> $$ \mathbb{L}_{SoftMax + TL} =  \mathbb{L}_{SoftMax} + \lambda \cdot \mathbb{L}_{TL} $$
> where
> $$ \mathbb{L}_{SoftMax} =  - log(\frac{e^{x_{class}}}{\sum_{i}e^{xi}}) $$
> and where $\lambda$ is a constant weighting hyperparameter and $\mathbb{L}_{TL}$ is standard triplet loss. For our experiments, we select $\lambda=0.01$ as suggested in the original paper as the result of a parameter grid search.
>
> This formulation is able to outperform the standard triplet loss approach since it combines the best of both worlds ‚Äì fully supervised learning and a separable embedded space. Most importantly for the task at hand, we propose to combine a fully supervised loss term as given by Softmax loss with the reciprocal triplet loss formulation which removes the necessity of specifying a margin parameter. This combination is novel and given by:
> $$ \mathbb{L}_{SoftMax + RTL} =  \mathbb{L}_{SoftMax} + \lambda \cdot \mathbb{L}_{RTL} $$
> --*Pg. 16-18, Andrew et al. (2020) [<sup>[1]</sup>](#fn1)*

Based on the above discussion, we implemented $\mathbb{L}_{SoftMax + TL}$ and the proposed $\mathbb{L}_{SoftMax + RTL}$ loss functions. 

> üìì **It is worth noting that the paper mentions an example mining strategy to find non-trivial triplets, which reportedly improves model performance and requires a lower number of training iterations. However, we have not implemented the example mining strategy**

````python
# Reciprocal Triplet loss
class RTL(nn.Module):
    def __init__(self):
        super(RTL, self).__init__()

    def d(self, anchor, instance):
        return (anchor - instance).pow(2).sum(1)

    def forward(self, negative, anchor, positive, prediction=None, target=None):
        return self.d(anchor, positive) + 1 / self.d(anchor, negative)


# Triplet loss and softmax loss combined
class TLSL(nn.Module):
    def __init__(self, TLmargin=0.5, _lambda=0.01):
        super(TLSL, self).__init__()
        self._lambda = _lambda
        self.TL = nn.TripletMarginLoss(margin=TLmargin)
        self.CE = nn.CrossEntropyLoss()

    def forward(self, negative, anchor, positive, prediction, target):
        return (self._lambda * self.TL(anchor, positive, negative).sum()) + self.CE(
            input=prediction, target=target - 1
        )


class RTLSL(nn.Module):
    def __init__(self, _lambda=0.01):
        super(RTLSL, self).__init__()
        self._lambda = _lambda
        self.RTL = RTL()
        self.CE = nn.CrossEntropyLoss()

    def forward(self, negative, anchor, positive, prediction, target):
        rtlLoss = self._lambda * self.RTL(anchor, positive, negative).sum()
        entropy = self.CE(input=prediction, target=target - 1)
        return rtlLoss + entropy
````

In our experiments, we trained the Resnet-50 model using $\mathbb{L}_{SoftMax + TL}$ and $\mathbb{L}_{SoftMax + RTL}$ loss function for every train/test split (i.e. we conducted 20 experiments in total).
* Since we changed the training strategy (`Resnet-50` was frozen), some hyperparameters had to be changed. These parameters are listed below in **bold**, along with a rationale for the choices.

| Parameter | Ours | Andrew et al. (2020)[<sup>[1]</sup>](#fn1) | Remark |
|---:|---:|---|---|
| **Learning rate** | $1e^{-1}$ | $1e^{-5}$ | We chose a higher learning rate because we were training just the last fully connected layers from scratch. The higher LR gave us better convergence. |

## Training routine
Figure 2 illustrates the training process.
1. The process starts with loading a batch of 16 images to feed-forward the Resnet-50 model. The model produces two outputs; (1) the 128-dimensional embeddings corresponding to the 16 images, and (2) a prediction of classes or "Cow IDs" for every image in a batch. 
2. Next, Using the model output, one of the two loss functions ($\mathbb{L}_{SoftMax + TL}$, or $\mathbb{L}_{SoftMax + RTL}$) are calcualted.
3. In the optimisation step, the stochastic gradient descent algorithm uses the computed loss value to adjust the model weight using the process called [backpropagation](https://brilliant.org/wiki/backpropagation/).

<center><img width="1000" alt="portfolio_view" src="https://github.com/Asheeshkrsharma/Identification-OpenCows/blob/main/staticAssets/TrainingProcess.svg?raw=1"><p>Figure 2: The training flow diagram</p></center>

4. Every second iteration or epoch, the model is evaluated on a test dataset to make sure that the model is not overfitting or underfitting. The model's overall accuracy is also calculated at this stage. It is worth noting that the model is not optimised during the evaluation.
5. To calculate the model's accuracy, we generate n 128-dimension embeddings using the images seen during the training.
6. A k-Nearest Neighbour algorithm is fitted using the corresponding labels and the generated embeddings. Next, the kNN fitted on the training data is used to predict the class labels of embeddings generated from the test dataset (i.e. on images previously unseen). Accuracy is calculated using the kNN predictions and the ground-truth labels as a percentage. 

> üìì The following two cells were executed for $\mathbb{L}_{SoftMax + TL}$ and $\mathbb{L}_{SoftMax + RTL}$ loss functions for every train-test split file, yielding 20 separate models. Since sharing all the models would take large storage space, we have consolidated the results in the `Convergence_results.csv` file. The csv file also contains the results obtained by Andrew et al. (2020)[<sup>[1]</sup>](#fn1).

<br>

> üìì Since calculating the accuracy on every second epoch is computationally intensive, we inferred the model on a small number of images. Therefore, the results obtained in the experiment are not meant for the final assessment of a model's accuracy. However, the results are sufficient to understand model convergence as an emergent trend.

````python
# Some utility functions are not directly implemented in this notebook to minimise clutter.
from utils.experiment import experiment, evaluate, inferAll

def calculateAccuracy(model, device, testDataLoader, trainDataLoader, Full=False):
    if Full == False:
        testEMBD, testLabels = inferAll(model, testDataLoader, device, Num=50)
        trainEMBD, trainLabels = inferAll(model, trainDataLoader, device, Num=300)
    else:
        # Num == 50000, will make an inference on all images in the dataset.
        print("Infering the test dataset")
        testEMBD, testLabels = inferAll(model, testDataLoader, device, Num=50000)
        print("Infering the train dataset")
        trainEMBD, trainLabels = inferAll(model, trainDataLoader, device, Num=50000)
    KNN = KNeighborsClassifier(n_neighbors=5, n_jobs=-4)
    KNN.fit(trainEMBD, trainLabels)
    predictions = KNN.predict(testEMBD)
    correct = (predictions == testLabels)
    accuracy = float(numpy.count_nonzero(correct)) / (len(testLabels) - 1)
    return accuracy
````

````python
# Get the device, prefer Cuda over CPU
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# Choose a loss function
lossFn = RTLSL() # Reciprocal Triplet + Softmax loss
# lossFn = TLSL() # Triplet + Softmax loss

# Prepare dataset, model, and the optimiser
device, model, testDL, trainDL, optimiser =  experiment(
    ResNetMod,
    "./datasets/OpenSetCows2020/splits/67-33.json",
    lRate=1e-1,
    decay=1e-7,
    momentum=0.0,
)

# To store the training history
history, canvas = hl.History(), hl.Canvas()
loss, numEpochs = 0, 200

modelPrefix = "RTLSL_67-33" # To save the model to disk

# Start training
model.train()
with tqdm(range(numEpochs), desc="Training epochs", position=0) as pbar:
    pbar.refresh()
    for epoch in range(numEpochs):
        # Prep data 
        anchors, positives, negatives, anchorLabel, negativeLabel = next(
            iter(trainDL)
        )
        anchors, positives, negatives = (
            anchors.to(device), positives.to(device), negatives.to(device)
        )
        anchorLabel, negativeLabel = (
            anchorLabel.view(len(anchorLabel)).to(device),
            negativeLabel.view(len(negativeLabel)).to(device)
        )

        # Calculate loss and Back propagate
        def backPropagate():
          optimiser.zero_grad()
          # The forward method returns three embeddings
          negativeEMBD, anchorEMBD, positiveEMBD, prediction = model(
              negatives, anchors, positives
          )
          loss = lossFn(
              negativeEMBD, anchorEMBD,
              positiveEMBD, prediction,
              torch.cat((anchorLabel, anchorLabel, negativeLabel), dim=0),
          )
          loss.backward()
          return loss

        loss = optimiser.step(backPropagate)

        # Every second iteration, calculate model accuracy, evaluate the loss function
        # on the test dataset. Save the experiment history
        if epoch % 2 == 0:
            VLoss = evaluate(model, testDL, device, RTLSL())
            accuracy = calculateAccuracy(model, device, testDL, trainDL)
            # accuracy = 0.
            history.log(epoch, TLoss=loss.data, VLoss=VLoss, KNNAcc=accuracy)       
            model.train()
            with canvas:
              canvas.draw_plot([history["TLoss"], history["VLoss"]])
              canvas.draw_plot([history["KNNAcc"]])
        pbar.update()
    pbar.refresh()
````

````python
# Weights trained on google collab
testDataset = OpenSetCows2020(
    0,  # Fold 0
    "./datasets/OpenSetCows2020/splits/67-33.json",
    split="test",
    transform=True,  # --------‚é§
    combine=True,  # -----------> Arguments taken as specified in the Paper.
    suppress_info=True,  # ---‚é¶
)
trainDataset = OpenSetCows2020(
    0,  # Fold 0
    "./datasets/OpenSetCows2020/splits/67-33.json",
    split="train",
    transform=True,  # --------‚é§
    combine=True,  # -----------> Arguments taken as specified in the Paper.
    suppress_info=True,  # ---‚é¶
)

testingDataLoader = DataLoader(testDataset, batch_size=16, num_workers=2, shuffle=False)
trainingDataLoader = DataLoader(trainDataset, batch_size=16, num_workers=2, shuffle=False)
accuracy = calculateAccuracy(model, device, testingDataLoader, trainingDataLoader, Full=True)
print("\n KNN Accuracy on test dataset: {}".format(accuracy * 100))

# Save model and training history with apropriate name
torch.save(model.state_dict(), f"./models/{modelPrefix}_{round(accuracy*100, 1)}.pth")
history.save(f"./models/{modelPrefix}_{round(accuracy*100, 1)}_history.pkl")
````
